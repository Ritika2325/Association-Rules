```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
```

```python
mymovies=pd.read_csv("C:/my_movies.csv")
```

```python
mymovies
```

```python
mymovies.shape
```

```python
mymovies.fillna("0",inplace = True)
```

```python
mymovies
```

```python
movies=mymovies.iloc[:,5:]
movies
```

```python
#Apriori Algorithm
frequent_itemsets=apriori(movies,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=0.2)
rules
```

```python
plt.scatter(rules.support,rules.confidence)
```

```python
rules1=association_rules(frequent_itemsets,metric="lift",min_threshold=0.7)
rules1
```

```python
plt.scatter(rules1.support,rules1.confidence)
```

```python
rules2=association_rules(frequent_itemsets,metric="lift",min_threshold=1)
rules2
```

```python
plt.scatter(rules2.support,rules2.confidence)
```
